**Question 8) How many games does the Wimbledon Winner of 2023 has to win to clinch the tournament**

```python
df = pd.read_excel('/content/drive/MyDrive/tylerudite-atp-match-data/atp_2023.xlsx')
winner_data = df[(df['Tournament'] == 'Wimbledon') & (df['Winner'] == 'Alcaraz C.')]
win_rounds = np.sum(winner_data['W1']) + np.sum(winner_data['W2']) + np.sum(winner_data['W3']) + np.sum(winner_data['W4']) + np.sum(winner_data['W5'])
loss_rounds = np.sum(winner_data['L1']) + np.sum(winner_data['L2']) + np.sum(winner_data['L3']) + np.sum(winner_data['L4']) + np.sum(winner_data['L5'])
total = win_rounds + loss_rounds
to_win = (win_rounds/total)*10
print(f'For Carlos Alcaraz to win Wimbledon in 2023, he has to win for {to_win} games out of 10 ')

df = pd.read_excel('/content/drive/MyDrive/tylerudite-atp-match-data/atp_2022.xlsx')
winner_data = df[(df['Tournament'] == 'Wimbledon') & (df['Winner'] == 'Alcaraz C.')]
win_rounds = np.sum(winner_data['W1']) + np.sum(winner_data['W2']) + np.sum(winner_data['W3']) + np.sum(winner_data['W4']) + np.sum(winner_data['W5'])
loss_rounds = np.sum(winner_data['L1']) + np.sum(winner_data['L2']) + np.sum(winner_data['L3']) + np.sum(winner_data['L4']) + np.sum(winner_data['L5'])
total = win_rounds + loss_rounds
to_win = (win_rounds/total)*10
print(f'Carlos Alcaraz in Wimbledon 2022, won {to_win} games out of 10 ')
```
